```python
# üì¶ Librer√≠as necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci√≥n visual
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# üì• 1. Extracci√≥n de datos desde GitHub (JSON)
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"
df = pd.read_json(url)

# Verifica las columnas anidadas (listas o diccionarios)
columnas_invalidas = []
for col in df.columns:
    if df[col].apply(lambda x: isinstance(x, (dict, list))).any():
        columnas_invalidas.append(col)

# Elimina columnas con estructuras anidadas que no son √∫tiles directamente
df = df.drop(columns=columnas_invalidas)

# üßπ 2. Transformaci√≥n: limpieza de nombres de columnas
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_') \
             .str.replace('¬ø', '').str.replace('?', '')

# Visualiza columnas disponibles despu√©s de limpiar
print("üßæ Columnas del DataFrame:")
print(df.columns.tolist())

# üîç 3. Identificar la columna de churn autom√°ticamente
posibles_churn = [col for col in df.columns if 'churn' in col or 'baja' in col or 'cancel' in col or 'termin' in col]
print("\nüîé Posibles columnas de evasi√≥n (churn):", posibles_churn)

# Asignar la primera encontrada como columna churn
churn_column = posibles_churn[0] if posibles_churn else None

if churn_column is None:
    raise ValueError("No se encontr√≥ una columna que represente 'churn' o evasi√≥n del cliente.")
else:
    print(f"\n‚úÖ Se utilizar√° la columna '{churn_column}' como indicador de evasi√≥n.")

# üìä 4. Distribuci√≥n de evasi√≥n
sns.countplot(x=churn_column, data=df)
plt.title('Distribuci√≥n de Clientes (Evasi√≥n vs Permanencia)')
plt.xlabel('¬øCliente se dio de baja?')
plt.ylabel('Cantidad de Clientes')
plt.show()

# üßπ 5. Valores nulos y duplicados
print("\nValores nulos por columna:")
print(df.isnull().sum())

print("\nFilas duplicadas:", df.duplicated().sum())

# üìê 6. Estad√≠sticas descriptivas
print("\nResumen estad√≠stico:")
print(df.describe(include='all'))

# üìä 7. Evasi√≥n por tipo de contrato (si existe)
if 'tipo_de_contrato' in df.columns:
    sns.countplot(x='tipo_de_contrato', hue=churn_column, data=df)
    plt.title('Evasi√≥n por Tipo de Contrato')
    plt.xticks(rotation=45)
    plt.show()

# üíµ 8. Crear columna de facturaci√≥n diaria (si existe)
if 'facturacion_mensual' in df.columns:
    df['cuentas_diarias'] = df['facturacion_mensual'] / 30
    print("\nüìà Columna 'cuentas_diarias' creada exitosamente.")

# üìä 9. Total gastado vs evasi√≥n (si existe)
if 'total_gastado' in df.columns:
    sns.boxplot(x=churn_column, y='total_gastado', data=df)
    plt.title('Total Gastado vs Evasi√≥n')
    plt.show()

# üîó 10. Correlaci√≥n entre variables num√©ricas
corr_matrix = df.select_dtypes(include=np.number).corr()

if corr_matrix.empty:
    print("‚ö†Ô∏è No hay columnas num√©ricas disponibles para calcular correlaci√≥n.")
else:
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
    plt.title('Matriz de Correlaci√≥n')
    plt.show()
```
